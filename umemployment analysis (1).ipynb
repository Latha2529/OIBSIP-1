# ANALYZING EMPLOYMENT RATE AND UNEMPLOYMENT RATE

```python
df.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=df)
plt.show()
```

```python
df.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=df)
plt.show()
```

# UNEMPLOYMENT BASED ON STATES

```python
group = df.groupby(['States'])['Estimated Unemployment Rate'].max()
group
```

```python
group = df.groupby(["States"])["Estimated Unemployment Rate"].aggregate("max").reset_index()
plt.figure(figsize=(12,8))
sns.pointplot(group['States'].values, group['Estimated Unemployment Rate'].values, alpha=0.8)
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Highest Rates of Unemployment by States", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

# UNEMPLOYMENT BASED ON REGION 

```python
group = df.groupby(['Region'])['Estimated Unemployment Rate'].max()
group
```

```python
group = df.groupby(["Region"])["Estimated Unemployment Rate"].aggregate("max").reset_index()
plt.figure(figsize=(12,8))
sns.pointplot(group['Region'].values, group['Estimated Unemployment Rate'].values, alpha=0.8)
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Highest Rates of Unemployment by States", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Unemployment rate by Year
group = df.groupby(["Date"])["Estimated Unemployment Rate"].aggregate("max").reset_index()
plt.figure(figsize=(12,8))
sns.pointplot(group['Date'].values, group['Estimated Unemployment Rate'].values, alpha=0.8)
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.title("Highest Rates of Unemployment by Year", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```
